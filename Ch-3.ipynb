## Adding a new Feature

```python
room_length=[18,20,10,12,18,11]
r_breadth=[20,20,10,11,19,10]
r_type=['Big','Big','N','N','Big','N']
data=pd.DataFrame({'Length':room_length,'Breadth':r_breadth,'Type':r_type})
data
```

```python
data['Area']=data['Length']*data['Breadth']
data
```

## np.where()

```python
area=[4720,2430,1203,4561,1205]
price=[236,256,300,500,230]
data4=pd.DataFrame({'Area':area,'Price':price})
data4
```

```python
data4['Price']=np.where(data4['Price']>=300,'High',np.where(data4['Price']<200,'Low','Medium'))
data4
```
